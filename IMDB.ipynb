```python
import pandas as pd
import seaborn as sns

```

```python
data= pd.read_csv('Netflix.csv')
data.head(10)
```

```python
data.dtypes
```

```python
df1=data[data['release_year'].between(2000,2024)].sort_values(by=['imdb_score','imdb_votes']).reset_index()
#df1.drop(columns=['level_0','index','id'])
df2=df1.drop(['level_0','index','id','age_certification'],axis=1)
df2.rename(columns={'title':'Name'})
```

```python
#Find Show
show = df2[df2['type']=='SHOW']
show
print('Total tv show 2000 to 2024:-',show.shape[0])
```

```python
#Highest imdb score 2000 to 2022
max_1=show[show['imdb_score']==show['imdb_score'].max()]
max_1.rename(columns={'title':' Show Name'}).sort_values(by='release_year')
```

```python
#lowest imdb score 2000 to 2022
low_1 =show[show['imdb_score']==show['imdb_score'].min()]
low_1.rename(columns={'title':' Show Name'})
```
